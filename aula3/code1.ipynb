```python
import numpy as np
from PIL import Image  
from numpy import asarray
import matplotlib.pyplot as plt
```

```python
def main ():
    # Open image
    image = Image.open('Images/lena.jpg')

    #image.show()
    #plt.imshow(image)

    #convert to numpy array
    npImage = np.array(image)
    #show image
    plt.imshow(npImage, cmap='gray')
    #plt.show()
    
    #operacao pixel a pixel
    darkImage = npImage.copy()
    darkImage = darkImage / 5
    
    #create two columns plot
    # plt1 = plt.subplot(1,2,1)
    # plt2 = plt.subplot(1,2,2)
    # plt1.title.set_text('Original image')
    # plt2.title.set_text('Dark image')
    
    # plt1.imshow(npImage, cmap="gray")
    # plt2.imshow(darkImage, cmap="gray", vmin=0, vmax=255)
    # plt.show()
    compareImages(npImage, darkImage)
```

```python
def compareImages(originalImage, filter):
    npOriginalImage = np.array(originalImage)
    plt1 = plt.subplot(1,2,1)
    plt2 = plt.subplot(1,2,2)
    
    plt1.title.set_text('Original')
    plt2.title.set_text('Updated')
    
    plt1.imshow(npOriginalImage, cmap="gray")
    plt2.imshow(filter, cmap="gray", vmin=0, vmax=255)
    
    
```

```python
if __name__ == "__main__":
    main()
```
