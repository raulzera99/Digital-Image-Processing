```python
import numpy as np
from PIL import Image  
from numpy import asarray
import matplotlib.pyplot as plt
```

```python
def main ():
    im = Image.open("Images/lena_gray_512_salt_pepper.tif")
    f_image = np.array(im)
    g_image = np.zeros(f_image.shape) # media
    h_image = np.zeros(f_image.shape) # mediana
    
    # Neighbourhood operation (operacao de vizinhanca)
    # line 
    l = f_image.shape[0]
    # column
    c = f_image.shape[1]
    # kernel size - size of neighbourhood
    k = 3
    
    for x in range (k, l-k):
        for y in range(k, c-k):
            # # get the neighbourhood
            # neighbourhood = f_image[x-k:x+k+1, y-k:y+k+1]
            # # get the median value
            # median = np.median(neighbourhood)
            # # set the median value
            # f_image[x,y] = median
            s_xy = f_image[x-k:x+k+1, y-k:y+k+1]
            g_image[x,y] = np.mean(s_xy).astype(int)
            h_image[x,y] = np.median(s_xy).astype(int)
            
            
    plt1 = plt.subplot(1,3,1)
    plt2 = plt.subplot(1,3,2)
    plt3 = plt.subplot(1,3,3)
    
    plt1.title.set_text('Original')
    plt2.title.set_text('Media')
    plt3.title.set_text('Mediana')
    
    plt1.imshow(f_image, cmap="gray")
    plt2.imshow(g_image, cmap="gray", vmin=0, vmax=255)
    plt3.imshow(h_image, cmap="gray", vmin=0, vmax=255)
    
    
    #compareImages(f_image, g_image)
    
```

```python
def compareImages(originalImage, filter):
    npOriginalImage = np.array(originalImage)
    plt1 = plt.subplot(1,2,1)
    plt2 = plt.subplot(1,2,2)
    
    plt1.title.set_text('Original')
    plt2.title.set_text('Updated')
    
    plt1.imshow(npOriginalImage, cmap="gray")
    plt2.imshow(filter, cmap="gray", vmin=0, vmax=255)
```

```python
if __name__ == '__main__':
    main()
```
