```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
```

```python
#código que gera imagem com as iniciais do nome RAUL geradas em numpy.
#amostragem é a digitalização dos valores das coordenadas x e y da imagem.
#quantização é a digitalização dos valores de amplitude de cor da imagem.


def generate_letter_image(letter, font_path, image_size=(100,100), font_size=60, background_color=(255, 255, 255), text_color=(0, 0, 0)):
    image = Image.new('RGB', image_size, background_color)
    draw = ImageDraw.Draw(image)
    
    try:
        font = ImageFont.truetype(font_path, font_size)
    except IOError:
        raise IOError(f"A fonte '{font_path}' não pode ser carregada. Verifique o caminho da fonte.") 
    
    text_width, text_height = draw.textsize(letter, font=font)
    x = (image_size[0] - text_width) // 2
    y = (image_size[1] - text_height) // 2
    
    draw.text((x, y), letter, fill=text_color, font=font)
    
    return image
```

```python
if __name__ == '__main__':
    letter_to_generate = "R"
    
    for letter_to_generate in "MURILO":
    
        #"NotoSansJP-VariableFont_wght.ttf"  
        font_path = "BebasNeue-Regular.ttf"
        #font_path = "NotoSansJP-VariableFont_wght.ttf"
        
        image = generate_letter_image(letter_to_generate, font_path)
        
        plt.imshow(image, cmap='gray')
        
        plt.show()
    #print(image.shape)
    
    #image.save(f"imgs/letters/{letter_to_generate}.png")
```
