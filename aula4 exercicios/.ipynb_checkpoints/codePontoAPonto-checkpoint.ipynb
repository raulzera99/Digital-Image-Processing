```python
import numpy as np
from PIL import Image  
from numpy import asarray
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

Calcular o negativo das imagens;</br>
Diminuir pela metade a intensidade dos pixels;</br>
Incluir 4 quadrados brancos 10 x 10 pixels em cada canto das imagens;</br>
Incluir 1 quadrado preto 15X15 no centro das imagens

```python
def negativeImg(img):
    img = asarray(img)
    img = 255 - img
    return img
```

```python
def halfIntesity(img):
    img = img/2
    return img
```

```python
#quadrado branco de 50x50 em cada canto da imagem
def whiteSquare(img):
    img = asarray(img)
    img[0:50,0:50] = 255
    img[0:50, img.shape[1]-50:img.shape[1]] = 255
    img[img.shape[0]-50:img.shape[0], 0:50] = 255
    img[img.shape[0]-50:img.shape[0], img.shape[1]-50:img.shape[1]] = 255
    
    return img
```

```python
# Incluir 1 quadrado preto 50X50 no centro das imagens
def blackSquare(img):
    img = asarray(img)
    img[int(img.shape[0]/2)-25:int(img.shape[0]/2)+25, int(img.shape[1]/2)-25:int(img.shape[1]/2)+25] = 0
    return img
```

```python
# Calcular o negativo das imagens
def exercicioA():
    imgLena = Image.open('lena_gray_512.tif')
    f_imgLena = np.array(imgLena)
    
    imgCameraman = Image.open('cameraman.tif')
    f_imgCameraman = np.array(imgCameraman)
    
    imgHouse = Image.open('house.tif')
    f_imgHouse = np.array(imgHouse)
    
    print("Exercicio A: Negativo da imagem")
    print()
    print("Imagens originais")
    
    plt.title(label="Imagens originais", 
          loc="center") 
    
    plt1 = plt.subplot(1,3,1)
    plt2 = plt.subplot(1,3,2)
    plt3 = plt.subplot(1,3,3)
    
    plt1.title.set_text("Lena")
    plt2.title.set_text("Cameraman")
    plt3.title.set_text("House")
    
    plt1.imshow(f_imgLena, cmap='gray')
    plt2.imshow(f_imgCameraman, cmap='gray', vmin=0, vmax=255)
    plt3.imshow(f_imgHouse, cmap='gray', vmin=0, vmax=255)
    
    plt.show()
    plt.figure()
    
    print()
    print("Imagens negativas")
    plt4 = plt.subplot(1,3,1)
    plt5 = plt.subplot(1,3,2)
    plt6 = plt.subplot(1,3,3)
    
    plt4.title.set_text("Lena")
    plt5.title.set_text("Cameraman")
    plt6.title.set_text("House")
    
    plt4.imshow(negativeImg(f_imgLena), cmap='gray')
    plt5.imshow(negativeImg(f_imgCameraman), cmap='gray', vmin=0, vmax=255)
    plt6.imshow(negativeImg(f_imgHouse), cmap='gray', vmin=0, vmax=255)
    
    plt.show()
    
    
```

```python
# Diminuir pela metade a intensidade dos pixels
def exercicioB():
    imgLena = Image.open('lena_gray_512.tif')
    f_imgLena = np.array(imgLena)
    
    imgCameraman = Image.open('cameraman.tif')
    f_imgCameraman = np.array(imgCameraman)
    
    imgHouse = Image.open('house.tif')
    f_imgHouse = np.array(imgHouse)
    
    print("Exercicio B: Diminuir pela metade intensidade dos pixels")
    print()
    print("Imagens originais")

    plt1 = plt.subplot(1,3,1)
    plt2 = plt.subplot(1,3,2)
    plt3 = plt.subplot(1,3,3)
    
    plt1.title.set_text("Lena")
    plt2.title.set_text("Cameraman")
    plt3.title.set_text("House")
    
    plt1.imshow(f_imgLena, cmap='gray')
    plt2.imshow(f_imgCameraman, cmap='gray', vmin=0, vmax=255)
    plt3.imshow(f_imgHouse, cmap='gray', vmin=0, vmax=255)
    plt.show()
    
    plt.figure()
    
    print("Imagens com metade da intensidade")
    plt4 = plt.subplot(1,3,1)
    plt5 = plt.subplot(1,3,2)
    plt6 = plt.subplot(1,3,3)
    
    plt4.title.set_text("Lena")
    plt5.title.set_text("Cameraman")
    plt6.title.set_text("House")
    
    plt4.imshow(halfIntesity(f_imgLena), cmap='gray')
    plt5.imshow(halfIntesity(f_imgCameraman), cmap='gray', vmin=0, vmax=255)
    plt6.imshow(halfIntesity(f_imgHouse), cmap='gray', vmin=0, vmax=255)
    
    plt.show()
    print()
```
